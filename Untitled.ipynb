## Finding the Optimal Number of Clusters (Elbow / Knee Method)

```python
#Extracting Features
def preprocess_data(data):
    # Scale the numerical features
    scaler = StandardScaler()
    numerical_features = data.columns[:-2]  
    data[numerical_features] = scaler.fit_transform(data[numerical_features])
    
    return data

#Finding the Optimal Number of Clusters
from sklearn.cluster import KMeans

def determine_optimal_clusters(data):
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    plt.plot(range(1, 11), wcss)
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()
# Preprocess the data
data = preprocess_data(data)
# Determine the optimal number of clusters
data_scaled = data.drop(['EC1', 'EC2'], axis=1)
determine_optimal_clusters(data_scaled)
```

## Model Training & Prediction

```python
# Fit K-means clustering (replace 'n_clusters' with the desired number of clusters)
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
clusters = kmeans.fit_predict(data_scaled)

# Add the cluster labels to the original data
data['Cluster'] = clusters

# Visualize the clusters 
plt.figure(figsize=(10, 10))
sns.scatterplot(data=data, x='MinEStateIndex', y='HallKierAlpha', hue='Cluster', palette='inferno')  # Replace 'MinEStateIndex' and 'HallKierAlpha' with the another features we want to visualize
plt.title('Clusters')
plt.xlabel('MinEStateIndex')
plt.ylabel('HallKierAlpha')
plt.show()
```

# Splitting & Preprocessing Data

```python
pr_data= data.drop(['Cluster'],axis=1)
data_E1=pr_data.drop(['EC2'],axis=1)
data_E2=pr_data.drop(['EC1'],axis=1)
```

```python
scaler = StandardScaler()
X_E1 = scaler.fit_transform(data_E1.drop(['EC1'], axis=1))
X_E2 = scaler.fit_transform(data_E2.drop(['EC2'], axis=1))
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and test sets for each target
X_train_EC1, X_test_EC1, y_train_EC1, y_test_EC1 = train_test_split(X_E1, data_E1['EC1'], test_size=0.2, random_state=42)
X_train_EC2, X_test_EC2, y_train_EC2, y_test_EC2 = train_test_split(X_E2, data_E2['EC2'], test_size=0.2, random_state=42)
```

# Model Training & Predicting

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import plot_roc_curve
```

```python
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier, AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
```

```python
models = {
    
    "XGB":XGBClassifier(),
    "CB":CatBoostClassifier(verbose=False),
    "GBC":GradientBoostingClassifier(),
    "AC":AdaBoostClassifier(),
    "RF":RandomForestClassifier(),
}
```

```python
#Training model for EC1
for name, model in models.items():
    print(f'Using model: {name}')
    model.fit(X_train_EC1,y_train_EC1)
    y_pred = model.predict(X_test_EC1)
    print("Rou_Accuracy {0:.2f}%".format(100*roc_auc_score( y_pred,y_test_EC1)))
    print(confusion_matrix(y_pred, y_test_EC1))
    print(f'RMSE : {np.sqrt(mean_squared_error(y_test_EC1,y_pred))}')
    print("-------------------------------------------")
```

```python
GBC_clf=GradientBoostingClassifier(n_estimators=50, learning_rate=0.1)
GBC_model=  GBC_clf.fit(X_train_EC1,y_train_EC1)
y_pred = GBC_model.predict(X_test_EC1)
print("Rou_Accuracy {0:.2f}%".format(100*roc_auc_score(y_pred,y_test_EC1)))
print(plot_confusion_matrix(GBC_clf, X_test_EC1, y_test_EC1))
```

```python
print(classification_report(y_pred, y_test_EC1))
```

```python
plot_roc_curve(GBC_model,X_test_EC1,y_test_EC1)
```

```python
#Training model for EC2
for name, model in models.items():
    print(f'Using model: {name}')
    model.fit(X_train_EC2,y_train_EC2)
    y_pred2 = model.predict(X_test_EC2)
    print("Rou_Accuracy {0:.2f}%".format(100*roc_auc_score(y_pred2,y_test_EC2)))
    print(confusion_matrix(y_pred2, y_test_EC2))
    print(f'RMSE : {np.sqrt(mean_squared_error(y_test_EC2,y_pred2))}')
    print("-------------------------------------------")
```
